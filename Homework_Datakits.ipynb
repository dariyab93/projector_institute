```python
import seaborn as sns
import matplotlib.pyplot as plt


# Loading exmample datasets
iris = sns.load_dataset("iris")
```

```python
iris.sample(10)
```

# Cheatsheet for plots

## Scatter plot using Seaborn

### Good for categorical data 


```python
sns.FacetGrid(iris, hue='species', size=12) \
   .map(plt.scatter, 'sepal_length', 'sepal_width') \
   .add_legend()
plt.title('Scatterplot with Seaborn', fontsize=20)
plt.xlabel('Sepal Length', fontsize=18)
plt.ylabel('Sepal Width', fontsize=18)
```

## Density plots (histograms, density, distribution)

### Good for continuous variables


```python
fig = plt.figure(figsize=(14,10))
sns.distplot(iris['sepal_width'], hist=True, kde=True, 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4})
plt.title('Density Plot and Histogram of Sepal Width', fontsize=18)
plt.xlabel('Sepal Width', fontsize=18)
plt.ylabel('Density', fontsize=18)
```

## Box plots, violin plots, strip plots 

### Good for continuous and categorical variables. Plots distributions in each category 


```python
plt.figure(figsize=(14, 10), dpi=80)

ax = sns.boxplot(iris['species'], iris['sepal_width'], boxprops=dict(alpha=0.75))
ax = sns.stripplot(iris['species'], iris['sepal_width'], jitter=True, edgecolor="gray")

ax.tick_params(axis='x', labelsize=14)
ax.tick_params(axis='y', labelsize=14)

ax.set_xlabel("Species", fontsize=16)
ax.set_ylabel("Sepal Width", fontsize=16)

# set a title
ax.set_title("Strip Plot with a Box plot comparing sepal width in different classes", fontsize=14)
```
